```python
import os
import glob
import pandas as pd
import expt
import matplotlib.pyplot as plt
print("expt:", expt.__version__)
```

```python
runs = expt.get_runs("./tmp/*")
h = runs.to_hypothesis("test")

from expt.plot import GridPlot

g = GridPlot(y_names=["a", "b"])
h.plot(ax=g["a"], title="a",
            x='global_step', y='train/success',
            err_style='fill', std_alpha=0.15,
            rolling=50, n_samples=5000, legend=False,
           )
h.plot(ax=g["b"], title="b",
            x='global_step', y='train/success',
            err_style='fill', std_alpha=0.15,
            rolling=50, n_samples=9528, legend=False,
           )
g.fig;
```

```python
# plot each run
y_names=["a", "b", "c"]
g = GridPlot(y_names=y_names, layout=(1, 3))
for run, name in zip(h, y_names):
    run.plot(ax=g[name], x='global_step', y='train/success', rolling=50)
```
